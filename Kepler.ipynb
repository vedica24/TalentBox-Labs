```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dp['koi_pdisposition'] = label_encoder.fit_transform(dp['koi_pdisposition']) 
dp['koi_disposition'] = label_encoder.fit_transform(dp['koi_disposition']) 

dp.head()
```

```python
## droping un-useful columns and removing columns if the missing values are more than 200
dp = dp.drop(columns = ['kepid', 'kepoi_name', 'kepler_name', 'koi_tce_delivname'])
dp = dp.pipe(lambda df: df.loc[:, (df.isnull().sum(axis='rows') < 500).where(lambda v: v).dropna().index.values]).dropna()
```

```python
dp.loc[dp['koi_fpflag_nt'] == 465, 'koi_fpflag_nt'] = 1
```

```python
dp.head()
```

```python
y = dp.iloc[:,0]
X = dp.iloc[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=.40)
```

### PCA - Principal Component Analysis for feature selection

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)


from sklearn.decomposition import PCA
pca = PCA(.80)

pca.fit(X_train)


train_img = pca.transform(X_train)
test_img = pca.transform(X_test)
```

### Models - 
###  1.) Logistic Regression
### 2.) Decision Trees
### 3.) random forest classifier

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, precision_score, confusion_matrix, accuracy_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression(C = 0.01, max_iter = 100, penalty = 'l2')
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
print('Evaluation Metrics:')
print('Accuracy: ' + str(metrics.accuracy_score(y_test, y_pred)))
print('Recall: ' + str(recall_score(y_test, y_pred, average = 'micro')))
print('F1 Score: ' + str(metrics.f1_score(y_test, y_pred, average = 'micro')))
print('Precision: ' + str(metrics.precision_score(y_test, y_pred, average = 'micro')))
print('\nConfusion Matrix:')
print(' TN,  FP, FN, TP')
confusion_matrix(y_test, y_pred)
```
